### Spliting the DataFram into Chunks

#### Split a DataFrame into chunks (with equal pieces by removing the pieces that do not contain the same size as the others)
it's a CSV file that contains two columns and 365 rows, so we are going to split this DataFrame into chunks "15 rows per Chunk" and each peace will be equal to all the others pieces, the last chunk will be verified if it's equal to all the chunk, if it's not it will be droped, but you can use the "else" condition to conserve it and to get another chunks that will not equal the others chunks or pieces.

```python
max_rows = 15
df_divided = []
while len(df) > max_rows:
    top = df[:max_rows]
    df_divided.append(top)
    df = df[max_rows:]
''' else:
    validation_div.append(validation_df)'''
```

```python
#length of the DataFrame_365Days
print("length is ",len(df_divided))
print("..................................................................")
#to print just the values of the columns 'y'
for i in df_divided:
    k=i['y'].values
    print(k)
```
